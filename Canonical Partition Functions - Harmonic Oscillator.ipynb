```python
import numpy as np
import matplotlib.pyplot as plt
```

# Harmonic Oscillator

A quantum harmonic oscillator is the quantum version of a spring-mass system that oscillates around an equilibrium. Unlike a classical oscillator, its energy can take only discrete values given by $E_n \ = \ \hbar \omega(n + 1/2)$ where n is $n = 0, 1, 2, \cdots$

The partition function of the quantum harmonic oscillator is given by:

$$Z = \frac {e^{- \frac {1}{2} \beta \cdot w}}{1 - e^{-\beta \cdot w}}$$

```python
# --- parameters ---
delta = 1.0               # energy gap Δ (in energy units)
kB = 1.0                  # Boltzmann constant (set to 1 for dimensionless units)
                          # if you want SI units, set kB = 1.38064852e-23 and adjust delta accordingly.
T = np.linspace(0.01, 5.0, 100)   # temperature array (same units consistent with delta/kB)

omega = 1.0           # frequency (arbitrary units)
kB = 1.0              # Boltzmann constant (set to 1 for simplicity)

# convenience
beta = 1.0 / (kB * T)
x = beta * delta / 2.0    # argument used in cosh, tanh, sech

# Partition function (not used directly for plotting but useful)
Z = 2.0 * np.cosh(x)
```

## Internal Energy (U)

$$U = -\frac {d}{db} ln \ z$$
$$U = \frac {d}{db} ln \left( \frac {e^{- \frac {1}{2} \beta \cdot w}}{1 - e^{-\beta \cdot w}} \right)$$
$$U = \cdot w \left( \frac {1}{2} + \frac {1}{e^{\beta \cdot w} -1}\right)$$

```python
# Internal energy U
U = omega * (0.5 + 1 / (np.exp(beta * omega) - 1))

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, U)
plt.xlabel('Temperature')
plt.ylabel('Internal Energy U')
plt.title('Two-Level System — Internal Energy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Harmonic Oscillator - Internal Energy.png')
plt.show()
```

## Heat Capacity (C)

$$C_V = \left (\frac {\partial U}{\partial T} \right)_V$$
$$C_V = \frac{\partial}{\partial T} \left( \frac {1}{2} + \frac {1}{e^{\beta \cdot w} -1}\right)$$
$$C_V = k_B(\beta \cdot w)^2 \frac {e^{\beta \cdot w}}{(e^{\beta \cdot w} - 1)^2}$$

```python
# Heat capacity
Cv = kB * (beta * omega)**2 * np.exp(beta * omega) / ( (np.exp(beta * omega) - 1)**2 )

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, Cv)
plt.xlabel('Temperature')
plt.ylabel('Heat Capacity C')
plt.title('Two-Level System — Heat Capacity')
plt.grid(True)
plt.tight_layout()
plt.savefig('Harmonic Oscillator - Heat Capacity.png')
plt.show()
```

## Helmholtz Free Energy (F)

$$F = - k_B T \ ln z$$
$$F = - k_B T \ ln \left (\frac {e^{- \frac {1}{2} \beta \cdot w}}{1 - e^{-\beta \cdot w}})\right)$$
$$F = \frac {^{\cdot} w}{2} + k_B T \ln \left (1 - e^{-\beta \cdot w}\right)$$

```python
# Helmholtz free energy F = - kB T ln Z
F = (omega/2) + kB * T * np.log(1 - np.exp(-beta * omega))

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, F)
plt.xlabel('Temperature')
plt.ylabel('Helmholtz Free Energy F')
plt.title('Two-Level System — Helmholtz Free Energy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Harmonic Oscillator - Helmholtz Free Energy.png')
plt.show()
```

## Entropy (S)

$$S = \frac {U - F}{T}$$
$$S = $$
$$S = k_B \left (\frac {\beta \cdot w}{e^{\beta \cdot w} - 1}\right) - ln (1 - e^{-\beta \cdot w})$$

```python
# Entropy S
S = kB * (beta * omega) / (np.exp(beta * omega) - 1) - np.log(1 - np.exp(-beta * omega))

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, S)
plt.xlabel('Temperature')
plt.ylabel('Entropy S')
plt.title('Two-Level System — Entropy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Harmonic Oscillator - Entropy.png')
plt.show()
```
