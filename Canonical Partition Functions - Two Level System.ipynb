```python
import numpy as np
import matplotlib.pyplot as plt
```

# Two State System

A simple quantum system that can exist in only two distinct energy states, and at a given temperature, it can jump between them, like a particle being either in a ground state or an excited state.

$$z = 2 \ cosh \left(\frac{\beta \Delta}{2} \right)$$

### Remarks
Unlike the ideal gas, we cannot define the _pressure_, _volume_, enthalpy, and gibbs function of a two-level system meaningfully, unless we embed the system to a larger space.


```python
# --- parameters ---
delta = 1.0               # energy gap Δ (in energy units)
kB = 1.0                  # Boltzmann constant (set to 1 for dimensionless units)
                          # if you want SI units, set kB = 1.38064852e-23 and adjust delta accordingly.
T = np.linspace(0.01, 5.0, 400)   # temperature array (same units consistent with delta/kB)

# convenience
beta = 1.0 / (kB * T)
x = beta * delta / 2.0    # argument used in cosh, tanh, sech

# Partition function (not used directly for plotting but useful)
Z = 2.0 * np.cosh(x)
```

## Internal Energy

$$U = \frac {d}{db} ln \ z$$
$$U = - \frac{\partial}{\partial \beta} \ ln \left (2 \cosh \ \left (\frac {\beta \Delta}{2} \right)\right)$$
$$U = - \frac{\Delta}{2} \ tanh \left(\frac{\beta \Delta}{2} \right)$$

```python
# Internal energy U
U = - (delta / 2.0) * np.tanh(x)

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, U)
plt.xlabel('Temperature')
plt.ylabel('Internal Energy U')
plt.title('Two-Level System — Internal Energy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Two Level System - Internal Energy.png')
plt.show()
```

## Heat Capacity
$$C_V = \left (\frac {\partial U}{\partial T} \right)_V$$
$$C_V = \frac{\partial}{\partial T} \left (-\frac {\Delta}{2} \ tanh \left (\frac {\beta \Delta}{2}\right) \right)$$
$$C_V = K_B \left(\frac {\beta \Delta}{2}\right)^2 sech^2 \left(\frac{\beta \delta}{2}\right)$$

```python
# Heat capacity C
C = kB * (x**2) * (1.0 / np.cosh(x)**2)

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, C)
plt.xlabel('Temperature')
plt.ylabel('Heat Capacity C')
plt.title('Two-Level System — Heat Capacity')
plt.grid(True)
plt.tight_layout()
plt.savefig('Two Level System - Heat Capacity.png')
plt.show()
```

## Helmholtz Free Energy
$$F = - k_B T \ ln z$$
$$F = - k_B T \ ln \left (2 \cosh \ \left (\frac {\beta \Delta}{2} \right)\right)$$

```python
# Helmholtz free energy F
F = - kB * T * np.log(Z)

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, F)
plt.xlabel('Temperature')
plt.ylabel('Helmholtz Free Energy F')
plt.title('Two-Level System — Helmholtz Free Energy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Two Level System - Helmholtz Free Energy.png')
plt.show()
```

## Entropy
$$S = \frac {U - F}{T}$$
$$S = \frac {\Delta}{2T} \ tanh \left(\frac{\beta \Delta}{2}\right) + k_B T \ ln \left (2 \cosh \ \left (\frac {\beta \Delta}{2} \right)\right)$$

```python
# Entropy S
S = (U - F) / T

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, S)
plt.xlabel('Temperature')
plt.ylabel('Entropy S')
plt.title('Two-Level System — Entropy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Two Level System - Entropy.png')
plt.show()
```
