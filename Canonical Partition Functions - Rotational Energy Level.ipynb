```python
import numpy as np
import matplotlib.pyplot as plt
```

# Rotational Energy Level

### Remarks

```python
# --- parameters ---
delta = 1.0               # energy gap Δ (in energy units)
kB = 1.0                  # Boltzmann constant (set to 1 for dimensionless units)
                          # if you want SI units, set kB = 1.38064852e-23 and adjust delta accordingly.
T = np.linspace(0.01, 5.0, 400)   # temperature array (same units consistent with delta/kB)

# convenience
beta = 1.0 / (kB * T)
x = beta * delta / 2.0    # argument used in cosh, tanh, sech

# Partition function (not used directly for plotting but useful)
Z = 2.0 * np.cosh(x)
```

## Internal Energy

```python
# Internal energy U
U = - (delta / 2.0) * np.tanh(x)

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, U)
plt.xlabel('Temperature')
plt.ylabel('Internal Energy U')
plt.title('Rotational Energy Level  — Internal Energy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Rotational Energy Level - Internal Energy.png')
plt.show()
```

## Heat Capacity

```python
# Heat capacity C
C = kB * (x**2) * (1.0 / np.cosh(x)**2)

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, C)
plt.xlabel('Temperature')
plt.ylabel('Heat Capacity C')
plt.title('Rotational Energy Level — Heat Capacity')
plt.grid(True)
plt.tight_layout()
plt.savefig('Rotational Energy Level - Heat Capacity.png')
plt.show()
```

## Helmholtz Free Energy

```python
# Helmholtz free energy F
F = - kB * T * np.log(Z)

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, F)
plt.xlabel('Temperature')
plt.ylabel('Helmholtz Free Energy F')
plt.title('Rotational Energy Level — Helmholtz Free Energy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Rotational Energy Level - Helmholtz Free Energy.png')
plt.show()
```

## Entropy

```python
# Entropy S
S = (U - F) / T

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, S)
plt.xlabel('Temperature')
plt.ylabel('Entropy S')
plt.title('Rotational Energy Level — Entropy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Rotational Energy Level - Entropy.png')
plt.show()
```
