```python
import numpy as np
import matplotlib.pyplot as plt
```

# Classical Ideal Gas

A collection of many tiny particles that move randomly and independently in a container, bouncing off walls but not interacting with each other, so all their energy is kinetic. The Canonical Partition Function for an ideal gas of N identical particles is defined as

$$Z_N = \frac {1}{N!} \left (\frac {V}{\lambda_{th}^3}\right)^N \propto (VT^{3/2})^N$$

Since $\lambda_{th} \propto T^{-1/2}$ hence we can write

$$ln \ Z_N = N \ ln \ V + \frac {3N}{2} ln \ T + constants$$

```python
# --- parameters ---
delta = 1.0                         # energy gap Δ (in energy units)
kB = 1.0                            # Boltzmann constant (set to 1 for dimensionless units)
                                    # if you want SI units, set kB = 1.38064852e-23 and adjust delta accordingly.
N = 20                              # Number of Particles
T = np.linspace(0.01, 5.0, 100)     # temperature array (same units consistent with delta/kB)
V = 1e-3                            # volume in m^3
h  = 6.62607015e-34                 # Planck constant (J·s)
m  = 4.65e-26                       # Example: mass of N2 molecule (kg)
n = N / V                           # number density

# convenience
beta = 1.0 / (kB * T)
x = beta * delta / 2.0    # argument used in cosh, tanh, sech

# Partition function (not used directly for plotting but useful)
Z = 2.0 * np.cosh(x)
```

## Internal Energy

$$U = \frac {d}{db} ln \ Z_N$$
$$U = \frac {d}{db} \ln \left (\frac {1}{N!} \left (\frac {V}{\lambda_{th}^3}\right)^N\right)$$
$$U = \frac {3}{2}Nk_BT$$

```python
## Internal energy U
U = (3/2) * N * kB * T

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, U)
plt.xlabel('Temperature (K)')
plt.ylabel('Internal Energy U')
plt.title('Ideal Gas — Internal Energy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Ideal Gas - Internal Energy.png')
plt.show()
```

## Helmholtz Free Energy (F)

$$F = -k_BT \ ln \ Z_n$$
$$F = -k_BTN \ ln \ V - k_B \frac {3N}{2} T \ ln T - k_BT \times constants$$

```python
## Helmholtz free energy F
F = (-kB * T * N * np.log(V)) - (kB * ((3 * N)/2) * T * np.log(T) )- (kB * T * 100)

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, F)
plt.xlabel('Temperature (K)')
plt.ylabel('Helmholtz Free Energy F')
plt.title('Ideal Gas — Helmholtz Free Energy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Ideal Gas - Helmholtz Free Energy.png')
plt.show()
```

## Entropy (S)
$$S = \frac {U - F}{T}$$
$$S = \frac {3}{2}Nk_B + Nk_B \ ln \left(\frac {Ve}{N \lambda_{th}^3}\right)$$
$$S = Nk_B \ ln \left(\frac {Ve^{5/2}}{N \lambda_{th}^3} \right)$$
$$S = Nk_B \left[\frac {5}{2} - ln (n \lambda_{th}^3)\right]$$

```python
## Entropy
# Thermal wavelength
lambda_th = h / np.sqrt(2 * np.pi * m * kB * T)

# Sackur–Tetrode formula
S = N * kB * ( 2.5 - np.log(n * lambda_th**3) )

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, S)
plt.xlabel('Temperature (K)')
plt.ylabel('Entropy S')
plt.title('Ideal Gas — Entropy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Two Level System - Entropy.png')
plt.show()
```

## Pressure (P)

$$p = - \left (\frac {\partial F}{\partial V} \right)_T$$
$$p = \frac {Nk_BT}{V}$$
$$p = Nk_BT$$

```python
## Pressure S
P = (N * kB * T)/V

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, P)
plt.xlabel('Temperature (K)')
plt.ylabel('Pressure P')
plt.title('Ideal Gas — Pressure')
plt.grid(True)
plt.tight_layout()
plt.savefig('Ideal Gas - Pressure.png')
plt.show()
```

## Enthalpy (H)

$$H = U + pV$$
$$H = \frac {5}{2}Nk_BT$$

```python
## Enthalpy
H = 5/2 * N * kB * T

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, H)
plt.xlabel('Temperature (K)')
plt.ylabel('Enthalpy H')
plt.title('Ideal Gas — Enthalpy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Ideal Gas - Enthalpy.png')
plt.show()
```

## Gibbs Free Energy (G)
$$G = H - TS$$
$$G = \frac {5}{2}Nk_BT - Nk_BT \ln \left(\frac {Ve^{5/2}}{N \lambda_{th}^3}\right)$$
$$G = Nk_BT \ ln(n\lambda_{th}^3)$$

```python
## Gibbs Free Energy (G)
G = H - (T * S)

plt.figure(figsize=(8,6), dpi = 100)
plt.plot(T, G)
plt.xlabel('Temperature (K)')
plt.ylabel('Gibbs Free Energy G')
plt.title('Ideal Gas — Gibbs Free Energy')
plt.grid(True)
plt.tight_layout()
plt.savefig('Ideal Gas - Gibbs Free Energy.png')
plt.show()
```

## Derived Thermodynamic Quantities

## Heat capacity at constant volume (CV)
Measures how much energy is required to raise temperature at fixed volume.

## Heat capacity at constant pressure (CP)
Measures thermal response at fixed pressure; related to CV.

## Compressibility
How volume responds to pressure.

## Thermal expansion coefficient
How volume changes with temperature.
